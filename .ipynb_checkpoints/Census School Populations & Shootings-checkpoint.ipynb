```python
# Reference to CSV and reading CSV into Pandas DataFrame
csv_path_census = "Data/Census School Populations_1955-2016.csv"
census_school_populations_df = pd.read_csv(csv_path_census)
census_school_populations_df.head()
```

```python
# Rename the column headers
census_school_populations_df = census_school_populations_df.rename(columns={'Year': 'Year',
                                                          'Total enrolled': 'Total Students Enrolled',
                                                          'N-Total': 'Total: Nursery',
                                                          'N-Public': 'Total: Public Nursery',
                                                          'N-Private': 'Total: Private Nursery',
                                                          'K-Total': 'Total: Kindergarten',
                                                          'K-Public': 'Total: Public Kindergarten',
                                                          'K-Private': 'Total: Private Kindergarten',
                                                          'E-Total': 'Total: Elementary School',
                                                          'E-Public': 'Total: Public Elementary School',
                                                          'E-Private': 'Total: Private Elementary School',
                                                          'H-Total': 'Total: High School',
                                                          'H-Public': 'Total: Public High School',
                                                          'H-Private': 'Total: Private High School',
                                                          'C-Total': 'Total: College',
                                                          'C-Public': 'Total: Public College',
                                                          'C-Private': 'Total: Private College',
                                                          'C-Full time': 'Total: Full-time College'})

census_school_populations_df.head()
```

```python
# Reference to CSV and reading CSV into Pandas DataFrame
csv_path_shootings = "Data/School Shooting Data_1990-Present.csv"
school_shooting_df = pd.read_csv(csv_path_shootings)

school_shooting_df['Date'].astype(str)
school_shooting_df['Wounded'].astype(float)

school_shooting_df.drop(columns=['Source']).head()
```

```python
date = school_shooting_df['Date']

split_year = []

for i in date:
    year = i.split('-')
    year = year[0]
    split_year.append(year)
    
split_year
```

```python
school_shooting_df['Year'] = split_year

school_shooting_df.drop(columns=['Date']).head()
```

```python
school_shooting_df[school_shooting_df.Dupe != True]
new_school_shooting_df = school_shooting_df.drop(columns=['Dupe','Date','Source'])
new_school_shooting_df.head()
```
